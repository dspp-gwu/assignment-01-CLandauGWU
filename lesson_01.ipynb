### Loops

The last major concept in programming languages is loops. In Python, every loop is a for each loop, that is, there is some sort of iterable object over which you iterate.

    print("First loop")

      print(val)

    print("Second loop")
    for val in range(5):
      print(val)

    print("Third loop")
    for val in range(6, 10):
      print(val)
